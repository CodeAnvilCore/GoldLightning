```python
webdriver_name = 'Firefox'
# Valid time-ranges; aligned with yahoo finance options
valid_time_dict = {'one_year'    : '1_Y'
                           , 'five_year' : '5_Y'
                           , 'maximum'   : 'MAX'}
valid_webdriver = ['Chrome'
                           , 'Firefox']
        # Selenium webdriver mapping
webdriver_map = {'Chrome'     : ChromiumDriver
                              , 'Safari'   : WebDriver
                              , 'Firefox'  : FirefoxDriver
                              , 'IE'       : InternetExplorerDriver}
symbol = 'BHP.AX'
driver = webdriver_map[webdriver_name]()
url = f'https://finance.yahoo.com/quote/{symbol}/history/'
driver.get(url)
yahoo_attrib_ref =\
            'data-ylk'
date_range_button_yahoo_attrib_ref =\
            'elmt:menu;itc:1;elm:input;sec:qsp-historical;slk:date-select;subsec:calendar'
 # Find date range button
button_elements =\
    driver\
        .find_elements(by      = By.TAG_NAME
                        , value = 'button')
date_range_button = None
for button in button_elements:
    button_yahoo_attrib =\
        button.get_attribute(yahoo_attrib_ref)
    if button_yahoo_attrib == date_range_button_yahoo_attrib_ref:
        # date range button found
        date_range_button = button
if not date_range_button is None:
    # If date range button found; click it
    date_range_button.click()
 # Wait for date-range dialogue box to open
time.sleep(1)
# Select user nominated time-frame
new_button_elements =\
                driver\
                    .find_elements(by      = By.TAG_NAME
                                   , value = 'button')
time_range_select_button = None
for button in new_button_elements:
    button_value = button.get_attribute('value')
    if button_value == '5_Y':
        # Button with user nominated button found 
        time_range_select_button = button
if not time_range_select_button is None:
    # If date range button found; click it
    time_range_select_button.click()
```
